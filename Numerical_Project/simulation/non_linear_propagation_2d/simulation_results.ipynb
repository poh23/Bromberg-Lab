# Speckle pattern pump, 1mw Gaussian probe

## big speckle pattern, big probe

```python
from experiment_simulation_with_probe_2d import propagation_through_setup
import numpy as np
from Numerical_Project.utils.speckles.Speckles import Speckles
import matplotlib.pyplot as plt
from Numerical_Project.utils.heatmap_generator import XY_2d_heatmap

# Thin Cuvette measures 1.25x1.25x4.5 cm on the outside in set up
L = 1.25e-2 # length of the cuvette in meters
N = 2 ** 8 # resolution of propagating envelopes, usually n=2**9
square_width = 1.25e-3
d = 0.2  # distance to screen - 20cm
kerr_coefficient = -3e-10  # m^2/W

pump_amplitude = 5e1
probe_amplitude = 2.5e2
sigma_pump = 15e-5  # 50 microns
probe_lambda = 632e-9  # probe wavelength in meters
sigma_probe = 2.5e-4  # probe beam width
x = np.linspace(-0.5 * square_width, 0.5 * square_width, N)
x0 = 0
y0 = 0
# Create a meshgrid for the probe envelope
X, Y = np.meshgrid(x, x)  # Create a meshgrid for the coordinates

sys = Speckles(N, N, square_width, square_width)
sys.diffuser1_phase = sys.phase_mask_macro_pixels_method(4)
sys.slm_phase = np.zeros((N,N))

E_start = np.exp(-(sys.XX**2+sys.YY**2)/(sigma_pump)**2)

E_init = pump_amplitude*E_start
init_pump_envelope = sys.propagate_thin(E_init)

init_probe_envelope = probe_amplitude*np.exp(-((X - x0) ** 2 + (Y - y0) ** 2) / (sigma_probe ** 2))

propagation_through_setup(kerr_coefficient, (x, x), d, L, init_pump_envelope, init_probe_envelope, probe_lambda, prev_propagation_file_path="speckle_pump_high_amp_probe_2.5e2.npz")

```

## fourier transform of a speckle pattern with speckled phase

```python
from experiment_simulation_with_probe_2d import propagation_through_setup
import numpy as np
from Numerical_Project.utils.speckles.Speckles import Speckles
import matplotlib.pyplot as plt
from Numerical_Project.utils.heatmap_generator import XY_2d_heatmap, FFT_2d_heatmap
from Numerical_Project.utils.FFT.FFT_2D.fourier_transform_2d import fourier_transform_2d


N = 2 ** 8 # resolution of propagating envelopes, usually n=2**9
square_width = 1.25e-3
d = 0.2  # distance to screen - 20cm

pump_amplitude = 5e1
sigma_pump = 15e-5  # 50 microns
x = np.linspace(-0.5 * square_width, 0.5 * square_width, N)
x0 = 0
y0 = 0
# Create a meshgrid for the probe envelope
X, Y = np.meshgrid(x, x)  # Create a meshgrid for the coordinates

sys = Speckles(N, N, square_width, square_width)
sys.diffuser1_phase = sys.phase_mask_macro_pixels_method(4)
sys.slm_phase = np.zeros((N,N))

E_start = np.exp(-(sys.XX**2+sys.YY**2)/(sigma_pump)**2)
# plot initial pump envelope before diffuser
fig, axes = plt.subplots(3,2, figsize=(20, 15))
axes = axes.ravel()
XY_2d_heatmap(axes[0], fig, sys.XX, sys.YY, np.abs(E_start)**2, title="Pump intensity before diffuser", cmap="viridis")

E_init = pump_amplitude*E_start
init_pump_envelope = sys.propagate_thin(E_init)
XY_2d_heatmap(axes[1], fig, sys.XX, sys.YY, np.abs(init_pump_envelope)**2, title="Pump intensity after diffuser", cmap="viridis")
XY_2d_heatmap(axes[2], fig, sys.XX, sys.YY, np.angle(init_pump_envelope), title="Pump angle after diffuser", cmap="viridis")

dist = 0.1
k = 2 * np.pi / 532e-9  # wave number for 532 nm light
new_pump_envelope = init_pump_envelope * np.exp(1j*dist*k* np.abs(init_pump_envelope)**2)  # add speckled phase
XY_2d_heatmap(axes[3], fig, sys.XX, sys.YY, np.angle(new_pump_envelope), title="Pump angle after diffuser with speckled phase", cmap="viridis")
# Fourier transform of the speckled pump envelope
vx, vy, E_out_fur = fourier_transform_2d(new_pump_envelope, sys.XX[0, :], sys.YY[:, 0])
FFT_2d_heatmap(axes[4], fig, vx, vy, np.abs(E_out_fur)**2, title="Fourier transform of the speckled pump envelope intensity", cmap="viridis")
FFT_2d_heatmap(axes[5], fig, vx, vy, np.angle(E_out_fur), title="Fourier transform of the speckled pump envelope angle", cmap="viridis")


```

```python
from Numerical_Project.utils.speckles.Speckles import Speckles
from Numerical_Project.utils.FFT.FFT_2D.fourier_transform_2d import fourier_transform_2d
import matplotlib.pyplot as plt

DIFF1STRENGH = 2
N = 256
square_width = 1.25e-3

sys = Speckles(N, N, square_width, square_width)
sys.diffuser1_phase = sys.phase_mask_macro_pixels_method(2)
sys.slm_phase = np.zeros((N,N))

E_start = np.exp(-(sys.XX**2+sys.YY**2)/(sys.sig/5)**2)
E_init = E_start/np.sqrt(np.sum(np.abs(E_start**2)))
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

E_out = sys.propagate_thin(E_init)

imm = axes[0].imshow(np.abs(E_init)**2)
axes[0].set_title('E in')
fig.colorbar(imm, ax=axes[0],shrink=0.8)
imm = axes[1].imshow(np.abs(E_out)**2)
axes[1].set_title('speckle pattern')
fig.colorbar(imm, ax=axes[1],shrink=0.8)
fig.suptitle("speckle after thick diffuser")

x = sys.XX[0, :]
y = sys.YY[:, 0]
vx, vy, E_out_fur = fourier_transform_2d(E_out,x, y)

imm = axes[2].imshow(np.abs(E_out_fur)**2)
axes[2].set_title('speckle pattern 2')
fig.colorbar(imm, ax=axes[2],shrink=0.8)
fig.suptitle("speckle after thick diffuser")
plt.show()
```
