## Pump and Probe

### 1mw Gaussian pump, 1mw Gaussian probe with width of 5*$\sigma_{pump}$

```python
from experiment_simulation_with_probe_2d import propagation_through_setup, get_pump_intensity
import numpy as np

# Thin Cuvette measures 1.25x1.25x4.5 cm on the outside in set up
L = 1.25e-2 # length of the cuvette in meters
N = 2 ** 8 # resolution of propagating envelopes, usually n=2**9
square_width = 1.25e-3
d = 0.2  # distance to screen - 20cm
kerr_coefficient = -3e-10  # m^2/W
sigma_pump = 5e-6  # 50 microns

x = np.linspace(-0.5 * square_width, 0.5 * square_width, N)
X, Y = np.meshgrid(x, x)  # Create a meshgrid for the coordinates

pump_amplitude = 3e4
init_pump_envelope = pump_amplitude*np.exp(-(X ** 2 + Y ** 2) / (sigma_pump ** 2))

probe_amplitude = pump_amplitude
probe_lambda = 632e-9  # probe wavelength in meters
sigma_probe = 5*sigma_pump # probe beam width
x0 = 0
y0 = 0
init_probe_envelope = probe_amplitude*np.exp(-((X - x0) ** 2 + (Y - y0) ** 2) / (sigma_probe ** 2))

propagation_through_setup(kerr_coefficient, (x, x), d, L, init_pump_envelope, init_probe_envelope, probe_lambda)
```
